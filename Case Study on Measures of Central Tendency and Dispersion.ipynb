### 3. Who scores the most based on average.. 

#### w.r.t gender 

```python
#for math
dataset[['gender','math score']].groupby('gender').mean().sort_values('math score',ascending=False)
```

```python
#for reading score
dataset[['gender','reading score']].groupby('gender').mean().sort_values('reading score',ascending=False)
```

```python
#for writing score
dataset[['gender','writing score']].groupby('gender').mean().sort_values('writing score',ascending=False)
```

On calculating averages among students wrt gender female has greater mean in both reading and writing scores.
But males show greater mean value than females in math score.

*Female in reading and writin scores and male in math score have higher marks*

#### w.r.t test preparation course 

```python
#for math
dataset[['test preparation course','math score']].groupby('test preparation course').mean().sort_values('math score',ascending=False)
```

```python
#for reading score
dataset[['test preparation course','reading score']].groupby('test preparation course').mean().sort_values('reading score',ascending=False)
```

```python
#for writing score
dataset[['test preparation course','writing score']].groupby('test preparation course').mean().sort_values('writing score',ascending=False)
```

On calculating averages among students wrt test preparation course,we can get a clear idea that for all category of scores, those who have completed the course is having higher marks when compared to the students who haven't completed their course.

*So there is a clear cut advance for those who have completed the course,since they scored higher with great difference from others*

### 4. Scoring variations 

#### w.r.t gender 

```python
#for math
dataset[['gender','math score']].groupby('gender').std().sort_values('math score')
```

```python
#for reading score
dataset[['gender','reading score']].groupby('gender').std().sort_values('reading score')
```

```python
#for writing score
dataset[['gender','writing score']].groupby('gender').std().sort_values('writing score')
```

Computed standard deviation wrt gender shows almost same variation for both male and female in the case of reading and writing scores while there is a slight variation in math scores between male and female.

#### w.r.t test preparation course 

```python
#for math
dataset[['test preparation course','math score']].groupby('test preparation course').std().sort_values('math score')
```

```python
#for reading score
dataset[['test preparation course','reading score']].groupby('test preparation course').std().sort_values('reading score')
```

```python
#for writing score
dataset[['test preparation course','writing score']].groupby('test preparation course').std().sort_values('writing score')
```

Variation study done by finding standard deviation wrt test preparation course gives small variations.
But the point to be noted is that *variation in marks is certainly less for those who have completed the test preparation course* when compared to those who haven't completed their course.

### 5. Top 25% students based on  math score

In order to find top 25% students based on math score we can consider the quartile method.
Quartiles divide data into 4 equal parts and the best one-forth of total students will be the ones who are above the 3rd quartile mark, that is above 75 percentile.

```python
#to get the third quartile mark
dataset.describe()
```

Above table shows that, students having math score of 77 and above are the top 25% in the class wrt math score.

```python
#list of top 25% students as needed to award bonus points
toppers = dataset[dataset["math score"]>=77]

pd.set_option('display.max_rows',255)

toppers
```

*Above is the list of 255 students  who are eligible for bonus points based on math score*
