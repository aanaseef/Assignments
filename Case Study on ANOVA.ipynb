**3. Whether all zones generate the same amount of sales..**

Inorder to find out whether all zones are generating equal amount of sales, we shall use ANOVA Test.
We use **one-way ANOVA test** here, as the groups are just the sales generated in each zones and is affecting a dependent variable,i.e, Total sales.


#### Hypothesis formation:


**H0: All zones generate the same amount of sales.**

**H1: All zones doesn't generate same amount of sales.**

Significance level,α = 0.05

```python
df.head(3)
```

```python
F,p = stats.f_oneway(df['Zone - A'],df['Zone - B'],df['Zone - C'],df['Zone - D'])
print('F test statistic=',round(F,4),'\nP-value=',round(p,4))
```

```python
if p >= 0.05:
    print('Accept the Null Hypothesis')
else :
    print ('Reject the  Null Hypothesis')
```

From the above one-way ANOVA test performed,as the p-value is less than the significance level(α = 0.05), it is evident that we have to accept alternative hypothesis(H1).
- Conclusion:
> All zones doesn't generate same amount of sales.
