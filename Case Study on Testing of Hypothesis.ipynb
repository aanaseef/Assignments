### To clarify whether there is any increase in sales after stepping into digital marketing: 

#### Hypothesis formation:

**H0: There is no difference in sales after stepping into digital marketing**  
**H1: There are changes in sales after stepping into digital marketing**

```python
ttest,pval = stats.ttest_rel(df['Sales_before_digital_add(in $)'],df['Sales_After_digital_add(in $)'])
print("T-value =",ttest)
print("P-value =",pval)
if pval<0.05:
    print("reject null hypothesis, i.e. H0 is rejected and a change in sales is visible after stepping into digital marketing")
else:
    print("accept null hypothesis, i.e. H1 is accepted, which means a change in sales is visible after stepping into digital marketing")
```

- The values(statistical discription) and the t-test obtained above shows significant change in sales, which is increasing, after stepping into digital marketing.
- There is significant increase in sales after stepping into digital marketing .

### Checking dependency between the features “Region” and “Manager”:

```python
data1 = df.groupby(['Region','Manager']).agg(['mean','count'])
data1
```

```python
data2 = pd.crosstab(df['Region'],df['Manager'])
data2
```

```python
stat, p, dof, expected = chi2_contingency(data2)

print("The Test chi-square value is :",stat)
print("The P-Value is :",p)
print("The Degree of freedom is :",dof)
```

#### Hypothesis formation:

**H0: There is no dependency between the features “Region” and “Manager”**  
**H1: There are dependencies between the features “Region” and “Manager”**

```python
critical_val = chi2.ppf(q=1-0.05,df=dof)
print("The critical value is:",critical_val)


if stat > critical_val:
    print("Reject the Null Hypothesis")
elif stat < critical_val:
    print ("Unable to reject the Null Hypothesis")
```

 - In this chi-square test, unable to reject H0 means there isn't a relationship between “Region” and “Manager”.
- **Thus above test shows there is no significant dependency between “Region” and “Manager”.**
