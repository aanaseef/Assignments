## 5. Visualizations

```python
iris_columns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.title('Count of floral classes')
sns.countplot(iris_data['Classification'])
```

```python
plt.figure(figsize=(10,12))

plt.subplot(2,1,1)
plt.title('Comparison of PL and PW b/w the species')
sns.scatterplot(iris_data['PL'],iris_data['PW'],hue=iris_data['Classification'],s=30)

plt.subplot(2,1,2)
plt.title('Comparison of SL and SW b/w the species')
sns.scatterplot(iris_data['SL'],iris_data['SW'],hue=iris_data['Classification'],s=30)

plt.tight_layout()
```

```python
data1=iris_data.loc[0:49,:]
data2=iris_data.loc[50:99,:]
data3=iris_data.loc[100:149,:]
```

```python
plt.figure(figsize=(8,12))

plt.subplot(3,1,1)
sns.heatmap(data1.corr(),annot=True)
plt.xticks(rotation=40)
plt.title("Dimension of flower parts in iris-setosa")

plt.subplot(3,1,2)
sns.heatmap(data2.corr(),annot=True)
plt.xticks(rotation=40)
plt.title("Dimension of flower parts in iris-versicular")

plt.subplot(3,1,3)
sns.heatmap(data3.corr(),annot=True)
plt.xticks(rotation=40)
plt.title("Dimension of flower parts in iris-virginica")

plt.tight_layout()
```
