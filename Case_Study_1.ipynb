### 4. Top 5 cars based on their sales price

```python
sale1 = dataset['Model'].loc[0:],dataset['Sale Price'].iloc[0:]
sale1 = pd.DataFrame(sale1)
sale1 = sale1.T
convert_data = {'Sale Price':float}
sale1 = sale1.astype(convert_data)#converting sale price to float data type.
sale1.info()
```

```python
top_5 = sale1.sort_values('Sale Price',ascending=False).head(5)
top_5
```

Top 5 cars based on their sale price are Jetta III,A6,LX,MX-6 and ES.
